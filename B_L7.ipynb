## Intervalli di confidenza o p-value?

L'intervallo di confidenza di una stima fornisce un range di possibilita per il valore di interesse nella popolazione, non una dicotomia arbitraria basata exclussivamente sulla significativita statistica.

= Tuttavia, anche il valore di p-value , e dovrebbe essere utilizzato in aggiunta all 'intervallo di confidenza

- di preferenza, **entrambi dovrebbere essere presentati**.

- sarabbe invece ** sempre da evitare** la presentazione di un risultato nei semplici termini "non significativo (P>0.05)" o "significativo (P < 0.05)"

- L’uso e la presentazione degli intervalli di confidenza diventa ancora più
rilevante quando si è nell’ambito degli studi clinici in cui si vuole
confrontare l’effetto di un trattamento di interesse rispetto a un controllo,
per capire se l’effetto osservato, oltre a essere statisticamente
‘significativo’ può essere ritenuto anche ‘clinicamente’ significativo o
rilevante.

- Il p-value infatti da solo non dice nulla circa la dimensione dell’effetto:
differenze anche piccole e non rilevanti dal punto di vista clinico
potrebbero emergere come ‘significative’ solo perché la dimensione
campionaria considerata è elevata e quindi l’errore standard della stima
dell’effetto bassa. Qualsiasi differenza, anche la più piccola e irrilevante,
con un’adeguata numerosità campionaria può emergere come
‘statisticamente significativa’.

## Uguaglianza tra trattamenti:
- Le domande “esiste una differenza tra i due trattamenti?” e “i due
trattamenti hanno lo stesso effetto?” hanno, nella lingua comune, lo stesso
significato. Sono semplicemente formulate in modo diverso. Una risposta
negativa a una di queste due domande implica la risposta affermativa
all’altra.
- Nel linguaggio statistico questo invece non è mai vero.

- Non riuscire a dimostrare una differenza significativa degli esiti tra due
gruppi (non poter rifiutare $H_0$) non implica che i due gruppi posti a confronto siano tra loro uguali ($H_0$ è vera).

- Infatti ci sarà sempre qualche piccola differenza negli effetti, e per ogni
piccola differenza vera ($\delta$), si può stabilire una dimensione del campione
tale che, con elevata probabilità, l'ipotesi nulla H0 possa essere rifiutata.

- **Un’ipotesi nulla non potrà mai essere verificata, ma solo falsificata.**

# Studi di NON-inferiorita

- Per molte patologie sta diventando sempre più difficile sviluppare farmaci
più efficaci di quelli già in commercio.

- Per questo motivo le aziende farmaceutiche sono alla ricerca di nuovi
trattamenti che abbiano approssimativamente la stessa efficacia ma
dimostrino migliori per altri aspetti.

- Questa tendenza ha stimolato lo sviluppo di RCT che mirano a dimostrare
che un trattamento sperimentale non è inferiore al trattamento di controllo.
Tali trial sono chiamate **trial di non inferiorità**.


- L’obiettivo di un RCT di non-inferiorità è valutare se l’efficacia di un nuovo
trattamento è buona quanto quella di un trattamento di riferimento.

- l nuovo trattamento può infatti offrire altri vantaggi (dare meno effetti collaterali, avere una maggior facilità di somministrazione, migliorare la qualità della vita) che rendono conveniente dimostrare la sua non-inferiorità in termini di efficacia.

- In questi studi l’inferenza si basa su considerazioni di tipo clinico e, dal punto di vista statistico, sull’esame dell’intervallo di confidenza della differenza tra i due gruppi posti a confronto.

- **Risulterebbe sbagliato però prendere decisioni sulla base del fatto che
l’ipotesi nulla "assenza di effetto" ($\delta = 0$) non possa essere rifiutata.**

- Se in uno studio di non-inferiorità prendessimo una decisione sulla base
del fatto che la differenza osservata ‘non è statisticamente significativa’,
commetteremmo un grosso errore.

- Diremmo infatti che un trattamento, potenzialmente inferiore rispetto a un
altro, è in realtà ‘equivalente o non-inferiore’ a quello di riferimento,
ponendo a rischio i futuri pazienti.

- Non riuscire a dimostrare una differenza significativa degli esiti tra due
gruppi (non rifiutare $H_0$) non implica che i due gruppi posti a confronto
siano tra loro uguali ($H_0$ è vera).

- **L’assenza di evidenza contro l’ipotesi nulla non significa che
l’ipotesi nulla sia vera.**

- Spesso, il fatto di non riuscire a rifiutare l’ipotesi nulla è dovuta a una
bassa potenza dello studio (scarsa numerosità campionaria)

- Cosa fare in queste situazioni? Siccome è impossibile ‘dimostrare’ statisticamente che l’effetto di due trattamenti è identico, si dovrà decidere un range (clinico) in cui considereremo equivalenti i due effetti.

- Il limite estremo di questo margine diventa la nostra nuova ipotesi nulla. Vogliamo dimostrare che la differenza tra il trattamento ipotizzato essere non-inferiore e quello di riferimento non oltrepassi questo limite.

## Studi di bioequivalenza:

- Dopo la scadenza di un brevetto, le aziende farmaceutiche possono
produrre e commercializzare la versione generica del farmaco di ’marca’
a condizione che la domanda di approvazione del nuovo farmaco venga
approvata.

- Le aziende di farmaci generici devono dimostrare che il loro prodotto è
bioequivalente a quello di ‘marca’.

- L’FDA (ente regolatore dei farmaci negli USA) definisce la
bioequivalenza tra due farmaci come:

""**the absence of a significant difference** in the rate and extent to which
the active ingredient [..] in pharmaceutical equivalents or pharmaceutical
alternatives becomes available at the site of drug action when
administered at the same molar dose under similar conditions in an
appropriately designed study "

- Se due composti sono bioequivalenti, a tutti gli effetti ci si aspetta che
siano lo stesso farmaco.

- Non è necessario dimostrarne l’efficacia, in quanto già dimostrata
dall’azienda che ha per prima l’ha sviluppato.

- Gli studi di bioequivalenza mirano a confrontare un farmaco generico
con il farmaco commerciale originale, per dimostrare che hanno lo
"stesso" profilo farmacocinetico (PK).

- Altre volte si vuole valutare se modalità diverse di somministrazionedello
stesso farmaco (esempio intravenosa vs orale, una pastiglia al giorno a
dose X vs due pastiglie al giorno a dose X/2) hanno lo stesso profilo PK,
espresso dai parametri di farmacocinetica più comuni:

- $C_{max}, C_{min}, AUC$

- fig 23 + 24

- Anche in questo caso, non potendo verificare una perfetta uguaglianza,
negli studi di bioequivalenza si valuta se la differenza tra i parametri di
farmacocinetica delle due formulazioni sta all’interno di uno
specificato intervallo che va da $\delta_L$ a $\delta_U$

## Analisi di uno studio di bioequivalenza
- $\mu_1$: AUC medio osservato con famaco "generico"
- $\mu_1$: AUC medio osservato con farmaco di "marca"

- Poiché i parametri di farmacocinetica (come l’AUC) hanno spesso distribuzione che approssima una log-normale, si preferisce lavorare sui loro logaritmi invece che sul valore originale. La differenza $\delta$ su cui fare inferenza e' quindi la differenza di $log{\mu_2} - \log(\mu_1)$, equivalente a $log{\mu_2/\mu_2} = log{R}$

- $R$- rapporto tra medie: $\delta= \log{\mu_2/\mu_1}$

fig: 25

## regola per definire la bioequivalenza:

Gli enti regolatori (come l’FDA) chiedono che, per concludere che due farmaci
siano bioequivalenti, l'intervallo di confidenza al 90% del rapporto di una
misura di PK log-trasformata (AUC e / o Cmax) rientri completamente
nell'intervallo 80-125%.

- ffig 26 + 27

Le ipotesi da verificare in un test di equivalenza hanno la seguente forma:

$H_0: \delta =< \delta_L $ or $\delta >= \delta_U$

vs

$H_1: \delta_L < \delta < \delta_U$

## Disegno cross-over (a una sequenza)
- Negli studi di bioequivalenza, si utilizzano spesso disegno di studio crossover: ogni paziente riceve i due diversi farmaci durante due diversi periodi di tempo: i pazienti passano da un trattamento all'altro nel corso dello studio.

I parametri di farmacocinetica vengono confrontati all’interno dello stesso
paziente (in questo modo si elimina la variabilità tra soggetti, ottenendo stime
precise con numerosità campionarie ridotte)

## Disegno cross-over per la verifica dell’efficacia

- La ragione per considerare un disegno crossover quando si pianifica un RCT è che risulta più efficiente di un classico disegno parallelo: poiché si tiene sotto controllo la variabilità naturale del parametro di interesse esistente tra pazienti (variabilità intracluster), quando questa è elevata (come spesso accade negli studi clinici) sono necessari molti meno pazienti per raggiungere lo stesso livello di potenza statistica o
precisione della stima di interesse di un disegno in parallelo.
- Intuitivamente, questo sembra ragionevole perché ogni paziente funge da proprio
controllo appaiato.
- Ogni paziente riceve sia il trattamento A che il trattamento B (da un punto di vista logico, è quanto di più vicino alla logica controfattuale si possa avere: si valuta cosa succede sotto B non in un gruppo di pazienti con caratteristiche in media simili a quelle sottoposte ad A, ma sullo stesso individuo).

## Inferenza causale, modello controfattuale, confrontabilità

Teoricamente, per essere certi di una relazione causa-effetto
(inferenza causale), si dovrebbe;

=> Confrontare l’outcome in un soggetto trattato con A con quello che sarebbe
potuto succedere se lo stesso soggetto fosse stato trattato con B

Logica "confronttuale"

Nella pratica, negli RCT con disegno in parallelo, poiché non è possibile
confrontare situazioni ‘controfattuali’ (sarebbe infatti necessario poter
andare indietro nel tempo), si confronta l’outcome in un gruppo di
soggetti trattati con A con l’outcome osservato in un gruppo “sostituto”
(gruppo di controllo) che non è stato esposto (ma che poteva esserlo);

Una valida inferenza causale si basa sulla validità del gruppo
di controllo



## Disegno cross-over per la verifica dell’efficacia

Sebbene il concetto di paziente che funge da proprio controllo sia molto attraente dal punto di vista dell’inferenza causale, a causa di problemi di tipo pratico i disegni crossover sono raramente impiegati negli studi per valutare l’efficacia di una terapia.

Perché il disegno fornisca stime corrette dell’efficacia, la malattia dovrebbe essere cronica e stabile e i trattamenti posti a confronto non dovrebbero portare a guarigioni o a miglioramenti permanenti, ma solo alleviare (temporaneamente) la malattia.

Se il trattamento A cura il paziente durante il primo periodo, il trattamento B non avrà l'opportunità di dimostrare la sua efficacia quando il paziente passa a questo nel secondo periodo.

Pertanto questo tipo di disegno funziona solo per quelle condizioni croniche (es: asma) dove non esiste una cura efficace, e i trattamenti disponibili tentano di migliorare la qualità della vita.

## L'obiettivo di un RCT di bioequivalenza è determinare se le formulazioni farmaceutiche

sperimentali e standard producono livelli di concentrazione ematica equivalenti: non siamo interessati a sapere se il farmaco ‘cura’ (efficacia), non siamo quindi interessati a parametri di efficacia .

Somministrando entrambe le formulazioni a ciascun soggetto, lasciando il tempo
sufficiente tra una somministrazione e l’altra affinché il farmaco non sia più rintracciabile nel sangue, non c’è il pericolo che l’effetto del farmaco somministrato per primo si trascini nel periodo successivo di trattamento.

Per questo motivo, il disegno crossover è il disegno più utilizzato negli studi di bioequivalenza.
